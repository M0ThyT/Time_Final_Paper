```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from random import randint
import pandas as pd
from fake_useragent import UserAgent
import time
```

```python
#random sleep function, good tool against bot detection
def rnd_sleep(x,y):
	rndnr = randint(x,y)
	time.sleep(rndnr)

#path for the geckodriver
path_gecko = r'/usr/local/bin/geckodriver'
```

```python
profile = webdriver.FirefoxProfile()
options = Options()         
#options.headless = True #comment out if you want to see the browser
ua = UserAgent(verify_ssl=False)
#options.add_argument("--headless")
options.add_argument('--disable-infobars')
options.add_argument('--disable-extensions')
options.add_argument('--profile-directory=Default')
options.add_argument('--incognito')
options.add_argument('--disable-plugins-discovery')
options.add_argument('--start-maximized')
userAgent = ua.random
options.add_argument(f'user-agent={userAgent}')
```

```python
url = 'https://www.ranker.com/list/2021-nfl-injury-report/pedro-cerrano'

browser = webdriver.Firefox(options = options,firefox_profile=profile, executable_path = path_gecko ) 
wait = WebDriverWait(browser, 300)  

browser.get(url) # open url
```

```python
#create empty pandas dataframe with columns name, week, injury, position
df = pd.DataFrame(columns = ['Name', 'Week', 'Injury', 'Position'])

blocks = browser.find_elements_by_xpath('//li[@class = "blogItem_container__WRfTk blogItem_content__D40Pl blogItem_landscapeContent__WqX7L"]')
for block in blocks:
    name = block.find_element_by_xpath('.//h2[@class = "HeaderContainer_title__RZc6m"]').text
    #Time = block.find_element_by_xpath('.//span[@class = "edS-b"]').text
    Info = block.find_element_by_xpath('.//div[@class = "richText_container__p5dag"]').text
    #Position = block.find_element_by_xpath('.//div[@class = "richText_container__p5dag"]/span/p[2]').text[0:3]
    #print(Injury_Time)
    print('-------')
    Info_split = Info.split('\n')
    Week = Info_split[0][0:7]
    Injury = Info_split[1][8:]
    Position = Info_split[2][0:3]
    #create dataframe with name, week, injury, position
    df2 = pd.DataFrame([[name, Week, Injury, Position]], columns = ['Name', 'Week', 'Injury', 'Position'])
    #merge dataframe with df
    df = pd.concat([df, df2], ignore_index=True)

#save dataframe to csv
df.to_csv('NFL_injuries.csv', index = False)


   
    
```
